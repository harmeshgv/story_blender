```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text
```

```python
pdf_text = extract_text_from_pdf("seq2seq_MT_paper_2014.pdf")
print(pdf_text)
```

```python
import pdfplumber

def extract_text_pdfplumber(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Example usage
pdf_path = "seq2seq_MT_paper_2014.pdf"
pdf_plumber_txt = extract_text_pdfplumber(pdf_path)
print(pdf_plumber_txt)
```

```python
import fitz  # PyMuPDF

def extract_text_pymupdf(pdf_path):
    document = fitz.open(pdf_path)
    text = ''
    for page in document:
        text += page.get_text()
    return text

# Example usage
pdf_path = "seq2seq_MT_paper_2014.pdf"
fitz_text = extract_text_pymupdf(pdf_path)
print(fitz_text)
```

```python
from pdf2image import convert_from_path
import pytesseract

def extract_text_ocr(pdf_path):
    images = convert_from_path(pdf_path)
    text = ''
    for image in images:
        text += pytesseract.image_to_string(image)
    return text

# Example usage
pdf_path = "seq2seq_MT_paper_2014.pdf"
ocr_text = extract_text_ocr(pdf_path)
print(ocr_text)
```

```python
from transformers import pipeline

def extract_characters_bert(script):
    nlp = pipeline("ner", grouped_entities=True)
    entities = nlp(script)
    characters = set(entity["word"] for entity in entities if entity["entity_group"] == "PER")
    return list(characters)

print(extract_characters_bert(pdf_text))
```

```python
import re
import spacy

def preprocess_script(script):
    # Remove scene headings and action descriptions
    script = re.sub(r"^(INT\.|EXT\.).*$", "", script, flags=re.MULTILINE)
    script = re.sub(r"^\s*\(.*\)\s*$", "", script, flags=re.MULTILINE)
    return script

def extract_characters(script):
    # Rule-based extraction
    characters = set(re.findall(r"^\s*([A-Z][A-Z\s]+)\s*$", script, re.MULTILINE))
    
    # NER-based extraction
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(script)
    characters.update(ent.text for ent in doc.ents if ent.label_ == "PERSON")
    
    return list(characters)

# Example usage

script = preprocess_script(pdf_text)
characters = extract_characters(pdf_text)
print(characters)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage you want to scrape
url = 'https://www.imdb.com/title/tt6263850/fullcredits/?ref_=tt_ov_ql_1'  # Replace with the actual URL

# Fetch the HTML content of the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table with class 'cast_list'
    cast_table = soup.find('table', class_='cast_list')
    
    # Initialize a list to store character names
    character_names = []
    
    # Iterate over each row in the table
    for row in cast_table.find_all('tr', class_=['odd', 'even']):
        # Find the 'td' with class 'character'
        character_td = row.find('td', class_='character')
        if character_td:
            # Extract all character names within the 'td'
            for character in character_td.find_all('a'):
                character_names.append(character.get_text(strip=True))
    
    # Print the list of character names
    print(character_names)
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import re

def extract_character_names(script_text):
    # Regex pattern to match character names in all caps (with optional colon or newline)
    pattern = r'^\s*([A-Z][A-Z\s]+)(?::|\n)'
    
    # Find all matches in the script text
    matches = re.findall(pattern, script_text, re.MULTILINE)
    
    # Normalize names: Remove extra spaces and convert to uppercase
    normalized_names = [name.strip().upper() for name in matches]
    
    # Remove duplicates by converting to a set, then back to a list
    character_names = list(set(normalized_names))
    
    # Sort the character names for easier reading
    character_names.sort()
    
    return character_names

character_names = extract_character_names(pdf_text)

# Print the results
print("Character Names:")
for name in character_names:
    print(name)
```
